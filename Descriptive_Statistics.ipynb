Q1: Understanding Central Tendency (Easy)
A bakery tracks the daily sales of muffins (in dozens) over a week: [10, 12, 11, 15, 14, 13, 12].
What is the most representative value of their weekly sales, and why?

Answer

The most representative value of the bakery’s weekly muffin sales is the mean (average).
Calculation:
Mean
=
10+12+11+15+14+13+12/7 = 12.43 dozens

Why the mean?
The data has no extreme outliers.
Sales are fairly consistent across the week.
The mean reflects the overall daily sales level better than the mode or the median, which ignore some variation.
So, the most representative value of weekly sales is about 12.4 dozens of muffins per day.

Q4: Median in Real Life (Medium)
A car dealer notes the prices of used cars: [$8,000, $9,500, $10,200, $11,000, $50,000].
Why is the median a better measure than the mean in this case? Calculate the median.

Answer

Median , Values are already in ascending order so the Median is $10,200.

Why the median is better than the mean:
The price $50,000 is an outlier compared to the other cars.
The mean would be pulled upward by this unusually high price and would not reflect the typical car price.

So, the median price of the used cars is $10,200, and it is a more reliable measure than the mean in this situation.

Answer

Range = MAX - MIN
        40 - 25 = 15 Min
This shows there is a noticeable variation in how long the student takes each day.
While the range gives a quick idea of spread, it only uses the extreme values and does not show how the times are distributed in between.

Q7: Variance for Decision-Making (Medium)
Two delivery companies track delivery delays (in minutes).
Company A: variance = 6
Company B: variance = 15
Which company is more consistent, and why?

Answer

Company A is more consistent.

Variance measures how spread out the delivery delays are from the average.
A lower variance means delays are closer to the mean and less unpredictable.
Company A’s variance (6) is much lower than Company B’s variance (15), indicating more reliable and consistent delivery times.

Q8: Standard Deviation in Context (Hard)
A finance student compares the daily price fluctuations of two cryptocurrencies.
Coin A: standard deviation = $30
Coin B: standard deviation = $120
Which coin is riskier to invest in, and why?

Answer

Coin B is riskier to invest in.

Standard deviation measures volatility, i.e., how much prices fluctuate around the average.
A higher standard deviation means larger and more frequent price swings.
Coin B’s standard deviation ($120) is much higher than Coin A’s ($30), indicating greater price volatility.
